```python
# Cell 1: setup & load
!pip install seaborn
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# change this if your path is different (I removed accidental extra spaces)
data_dir = Path(r"C:\Users\prath\Downloads\Cyber AIML POC\transaction_dataset.csv")
csv_files = list(data_dir.glob("*.csv"))
print("CSV files found:", csv_files)

df = pd.read_csv(data_dir)
print("Loaded shape:", df.shape)
df.head()

```

```python
# Quick info about data
print("Columns:", df.columns.tolist())
print("\nBasic info:")
display(df.info())
display(df.describe().transpose())

# Fraud/Legit distribution
print("\nFLAG value counts:")
print(df['FLAG'].value_counts(dropna=False))

plt.figure(figsize=(5,3))
sns.countplot(x='FLAG', data=df)
plt.title('Fraud vs Legit Counts')
plt.show()
```

```python
# Convert FLAG to 0/1 cleanly
def map_flag(x):
    s = str(x).strip().lower()
    if s in ('1','fraud','fraudulent','true','yes'):
        return 1
    if s in ('0','legit','legitimate','normal','false','no'):
        return 0
    try:
        v = float(s)
        return int(v)
    except:
        return np.nan

df['FLAG'] = df['FLAG'].apply(map_flag).astype(int)

print("FLAG after mapping:")
print(df['FLAG'].value_counts())
```

```python
# Drop Address & Index if they exist
drop_cols = [c for c in ['Address','Index'] if c in df.columns]

# Drop columns with >50% missing values
missing_pct = df.isnull().mean()
drop_cols += missing_pct[missing_pct > 0.5].index.tolist()

print("Dropping columns:", drop_cols)
df = df.drop(columns=drop_cols, errors='ignore')
```

```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split

X = df.drop(columns=['FLAG'])
y = df['FLAG']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Fraud ratio in train:", y_train.mean())
print("Fraud ratio in test:", y_test.mean())
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rf = RandomForestClassifier(
    n_estimators=100,
    class_weight="balanced",
    random_state=42
)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Random Forest Results:")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
import joblib

# choose the better one (Random Forest usually works better for tabular)
best_model = rf
joblib.dump(best_model, "fraud_detection_model.pkl")
print("Model saved as fraud_detection_model.pkl")
```

```python
new_transactions = X_test.sample(5, random_state=42)
print("Simulated new transactions:\n", new_transactions)
```
